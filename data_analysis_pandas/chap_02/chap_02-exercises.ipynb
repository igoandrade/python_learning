4. Adicione uma coluna indicando se o terremoto aconteceu em um país ou estado dos EUA que está no Anel de Fogo. Use Alasca, Antártica (procure Antártica), Bolívia, Califórnia, Canadá, Chile, Costa Rica, Equador, Fiji, Guatemala, Indonésia, Japão, Ilhas Kermadec, México (tome cuidado para não selecionar Novo México), Nova Zelândia, Peru, Filipinas, Rússia, Taiwan, Tonga e Washington.

```python
ring_of_fire = [
    'Alaska', 'Antarctic', 'Bolivia', 'California', 'Canada', 'Chile', 'Costa Rica', 
    'Ecuador', 'Fiji', 'Guatemala', 'Indonesia', 'Japan', 'Kermadec Islands', 'Mexico',
    'New Zealand', 'Peru', 'Philippines', 'Russia', 'Taiwan', 'Tonga', 'Washington'
]
```

```python
df['is_ring_of_fire'] = [int(_ in ring_of_fire) for _ in df.parsed_place]
df.head()
```

5. Calcule o número de terremotos (_earthquakes_) na região do Anel de Fogo (_Ring of Fire_) e o número fora dessa região.

```python
df.loc[
    (df['is_ring_of_fire'] == 1) & (df['type'] == 'earthquake'), ['is_ring_of_fire']
].shape[0]
```

```python
df.loc[
    (df['is_ring_of_fire'] == 0) & (df['type'] == 'earthquake'),
].shape[0]
```
